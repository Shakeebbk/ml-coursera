# 2. Linear Regression with one variable

```python
# 2.1 Plotting the data
data = np.loadtxt(os.path.join('data', 'ex1data1.txt'), delimiter=',')
X = data[:, 0]
y = data[:, 1]

m = y.size
```

```python
plt.plot(X, y, 'ro', ms=10, mec='k')
plt.xlim(4,24)
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s');
```

### 2.2 Gradient Descent Functions
**Cost Function**
![cost function](./data/1.PNG)

```python
# Assignment part
def computeCost(X, y, theta):
    m = y.size
    J = 0
    
    h = X.dot(theta)

    J = ((1.0/(2.0*m)) * np.sum(np.square(np.subtract(h, y))))
    
    return J
```

```python
# Add a column of ones to X. The numpy function stack joins arrays along a given axis. 
# The first axis (axis=0) refers to rows (training examples) 
# and second axis (axis=1) refers to columns (features).
X = np.stack([np.ones(m), X], axis=1)

# Test out the cost function
c = computeCost(X, y, theta=np.array([0.0, 0.0]))
print ("With theta = [0, 0]\nCost computed = %f" % (c))
print ('Expected cost value (approx) 32.07\n')

c = computeCost(X, y, theta=np.array([-1, 2]))
print ("With theta = [-1, 2]\nCost computed = %f" % (c))
print ('Expected cost value (approx) 54.24\n')
```

**Gradient descent**
![gradient descent](./data/2.PNG)

```python
# Assignment part
def gradientDescent(X, y, theta, alpha, iterations):
    m = y.size
    J_history = np.zeros(iterations) # Storing historical costs for plotting and analysis
    
    for iter in range(0, iterations):
        h = X.dot(theta)
        
        theta = theta - (alpha / m) * X.T.dot(np.subtract(h, y)) # **REMEMBER** --> 1/m = 1.0/m for float considerations
        
        J_history[iter] = computeCost(X, y, theta)
    
    return (theta, J_history)
        
```

```python
theta = np.zeros(2) # Initializing fitting parameters

iterations = 1500
alpha = 0.01

theta, J_history = gradientDescent(X, y, theta, alpha, iterations)

print ('Theta found by gradient descent:\n');
print (theta.ravel());
print ('Expected theta values (approx)\n');
print ('[-3.6303 1.1664]\n');
```

```python
# visualising linear regression fit
fig = plt.figure()
plt.plot(X[:, 1], y, 'ro', ms=10, mec='k')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.plot(X[:,1], np.dot(X, theta), '-')
plt.legend(['Training Data', 'Linear Regression']);
```

```python
# Predictions
print ('For population = 35,000, we predict a profit of %f' % (np.dot([1, 3.5], theta) * 10000))
print ('For population = 70,000, we predict a profit of %f' % (np.dot([1, 7], theta) * 10000))
```

```python
# visualising J(theta_0, theta_1)
theta_0_vals = np.linspace(-10, 10, 100)
theta_1_vals = np.linspace(-1, 4, 100)

J_vals = np.zeros((len(theta_0_vals), len(theta_1_vals)))

# filprint J_vals
for i, theta0 in enumerate(theta_0_vals):
    for j, theta1 in enumerate(theta_1_vals):
        J_vals[i, j] = computeCost(X, y, [theta0, theta1])

# Because of the way meshgrids work in the surf command, we need to
# transpose J_vals before calling surf, or else the axes will be flipped
J_vals = J_vals.T

# surface plot
fig = plt.figure(figsize=(12, 5))
# subplot - 121 -> 1 row, 2 columns, 1st box selected
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(theta_0_vals, theta_1_vals, J_vals, cmap='viridis')
plt.xlabel('theta0')
plt.ylabel('theta1')
plt.title('Surface')

# contour plot
ax = fig.add_subplot(122)
# Plot J_vals as 15 contours spaced logarithmically between 0.01 and 100
plt.contour(theta_0_vals, theta_1_vals, J_vals, linewidths=2, cmap='viridis', levels=np.logspace(-2, 3, 20))
plt.xlabel('theta0')
plt.ylabel('theta1')
plt.plot(theta[0], theta[1], 'ro', ms=10, lw=2)
plt.title('Contour, showing minimum');
```

# 3. Linear Regression with Multiple Variables(Optional):
## 3.1 Feature Normalization

```python
# Load the data
data = np.loadtxt(os.path.join('data', 'ex1data2.txt'), delimiter=',')
X = data[:, :2]
y = data[:, 2]
m = y.size
```

```python
print '{:>25s}{:>25s}{:>10s}'.format('X[:, 0] size(sft)', 'X[:, 1] no of rooms', 'y')
print '-'*60
for i in range(10):
    print '{:^25.0f}{:^25.0f}{:10.0f}'.format(X[i, 0], X[i, 1], y[i])
```

```python
# Normalize the features
def featureNormalize(X):
    """
    Normalizes the features in X. returns a normalized version of X where
    the mean value of each feature is 0 and the standard deviation
    is 1. This is often a good preprocessing step to do when working with
    learning algorithms.
    
    Parameters
    ----------
    X : array_like
        The dataset of shape (m x n).
    
    Returns
    -------
    X_norm : array_like
        The normalized dataset of shape (m x n).
    """
    mu = np.mean(X)
    
    sigma = np.std(X)
    
    X_norm = np.divide(np.subtract(X, mu), sigma)
    
    return X_norm, mu, sigma
```

```python
# call featureNormalize on the loaded data
X_norm, mu, sigma = featureNormalize(X)

print 'Computed mean:', mu
print 'Computed standard deviation:', sigma
```

```python
# Add intercept term to X
X = np.concatenate([np.ones((m, 1)), X_norm], axis=1)

alpha = 0.01;
num_iters = 400;

theta = np.zeros(X.shape[1])

theta, J_history = gradientDescent(X, y, theta, alpha, num_iters)
```

```python
# Plot J_history
plt.plot(J_history)
plt.xlabel('Number of iterations')
plt.ylabel('Cost J');
```

```python
print 'theta computed from gradient descent:'
print theta.ravel();
print computeCost(X, y, theta);
```

```python
# Predict price for - house size = 1650, no of bedrooms = 3
in_features, mu, sigma = featureNormalize(np.array([1650, 3]))
in_features = np.append(np.ones((1, 1)), in_features)

price = np.dot(in_features.T, theta)

print 'Predicted price of a 1650 sq-ft, 3 br house\n (using gradient descent):\n $%f' % (price);
```

```python
# Solving using normal equation
# Load the data
data = np.loadtxt(os.path.join('data', 'ex1data2.txt'), delimiter=',')
X = data[:, :2]
y = data[:, 2]
m = y.size

# # Add intercept term to X
# X = np.concatenate([np.ones((m, 1)), X], axis=1)

def normalEqn(X, y):
    return (np.linalg.pinv(X.T.dot(X)).dot(X.T)).dot(y)

theta_n = normalEqn(X, y)
price = np.dot(np.array([1650, 3]).T, theta_n)

print 'Predicted price of a 1650 sq-ft, 3 br house\n (using normal equation):\n $%f' % (price);
```
